# StackAPI

#### Import the necessary libraries here:

```python
# your code here
from stackapi import StackAPI
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
stack_api = StackAPI('stackoverflow')
comments = stack_api.fetch('comments')
```

```python
comments
```

```python
questions = stack_api.fetch('questions', fromdate=1560758400, todate=1563350400)
```

```python
questions
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions_today = stack_api.fetch('questions', fromdate=1560722401, todate=1563350400, tagged='python', sort='votes', min=5)
```

```python
questions_today
```
